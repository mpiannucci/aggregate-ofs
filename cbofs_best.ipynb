```python
import fsspec
import cf_xarray
import xarray as xr
import datetime
import matplotlib.pyplot as plt
import numpy as np
```

```python
# open dataset as zarr object using fsspec reference file system and xarray
fs = fsspec.filesystem("reference", fo='s3://nextgen-dmac-cloud-ingest/nos/cbofs/nos.cbofs.fields.best.nc.zarr', remote_protocol='s3', remote_options={'anon':True}, target_protocol='s3', target_options={'anon':True})
m = fs.get_mapper("") 

ds = xr.open_dataset(m, engine="zarr", backend_kwargs=dict(consolidated=False), 
                      chunks={})
ds
```

```python
mask = ds[f'mask_{ds.zeta.cf["latitude"].name.split("_")[1]}'].cf.isel(time=0).squeeze(drop=True).cf.drop_vars('time')
mask[:-1,:] = mask[:-1,:].where(mask[1:,:] == 1, 0)
mask[:,:-1] = mask[:,:-1].where(mask[:,1:] == 1, 0)
mask[1:,:] = mask[1:,:].where(mask[:-1,:] == 1, 0)
mask[:,1:] = mask[:,1:].where(mask[:,:-1] == 1, 0)
mask
```

```python
da = ds.zeta.where(mask == 1)
da
```

```python
da.cf.isel(time=0).cf.plot(x='longitude', y='latitude', cmap='jet')
```
