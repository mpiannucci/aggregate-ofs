```python
import cf_xarray
import fsspec
import xarray as xr

# open dataset as zarr object using fsspec reference file system and xarray
fs = fsspec.filesystem(
    "reference", 
    fo='s3://nextgen-dmac-cloud-ingest/nos/ngofs2/nos.ngofs2.fields.best.nc.zarr', 
    remote_protocol='s3', 
    remote_options={'anon':True}, 
    target_protocol='s3', 
    target_options={'anon':True}
)
m = fs.get_mapper("") 

ds = xr.open_dataset(m, engine="zarr", backend_kwargs=dict(consolidated=False), 
                      chunks={}, drop_variables=['Itime2'])

ds
```

Add the [UGRID](https://ugrid-conventions.github.io/ugrid-conventions/#2d-triangular-mesh-topology) mesh information

```python
ds["fvcom_mesh"] = (
    (),
    0,
    {
        "cf_role": "mesh_topology",
        "topology_dimension": 2,
        "node_coordinates": "lon lat",
        "face_coordinates": "lonc latc",
        "face_node_connectivity": "nv",
        "face_face_connectivity": "nbe",
    },
)
ds
```
