```python
import pandas as pd
import ujson
```

Read the data in from json format and create a `pandas` dataframe from data retreived from: 

https://api.eia.gov/v2/electricity/retail-sales/data/?frequency=monthly&data[0]=price&facets[stateid][]=MA&facets[stateid][]=NH&start=2013-01&end=2023-08&sort[0][column]=period&sort[0][direction]=asc&sort[1][column]=stateid&sort[1][direction]=asc&offset=0&length=5000

```python
target_file = 'monthly_electricity_data_ma_nh'
data = ujson.load(open(f'./{target_file}.json'))['response']['data']

df = pd.DataFrame(data)
df
```

```python
ma_df = df[df['stateid'] == 'MA']
for sector in df['sectorid'].unique():
    ma_df[ma_df['sectorid'] == sector][['period', 'price']].to_csv(f'./monthly_electricity_price_ma_{sector}.csv', index=False)
```

```python
ma_df = df[df['stateid'] == 'NH']
for sector in df['sectorid'].unique():
    ma_df[ma_df['sectorid'] == sector][['period', 'price']].to_csv(f'./monthly_electricity_price_nh_{sector}.csv', index=False)
```
